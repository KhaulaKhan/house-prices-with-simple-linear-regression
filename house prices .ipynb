```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
house_prices =  pd.read_csv(r"SLR_house_prices.csv")
house_prices.head()
```

```python
house_prices.shape
```

```python
house_prices.info()
```

```python
area_mean = house_prices['area'].mean()
price_mean = house_prices['price'].mean()
```

```python
house_prices['area'] = np.where(house_prices['area'].isnull(), area_mean, house_prices['area'])
```

```python
house_prices['price'] = np.where(house_prices['price'].isnull(), price_mean, house_prices['price'])
```

```python
house_prices.info()
```

```python
X, y = house_prices[['area']], house_prices['price']
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x, train_y, test_y = train_test_split(X, y, test_size = 0.2)
```

```python
train_x.shape, train_y.shape, test_x.shape, test_y.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model1 = LinearRegression()
```

```python
model1.fit(train_x, train_y)
```

```python
y_pred = model1.predict(test_x)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(test_y, y_pred)
```

```python
plt.plot(range(0, len(test_y)), test_y)
plt.plot(range(0, len(y_pred)), y_pred)
```

```python
plt.plot(range(0, 10), test_y[0:10])
plt.plot(range(0, 10), y_pred[0:10])
```
